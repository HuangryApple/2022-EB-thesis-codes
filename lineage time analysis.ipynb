### N content of des

```python
des_n=pd.read_csv('des_n_dis.csv')
```

```python
des_n_clean=des_n.drop(columns=['Unnamed: 0'])

n1=des_n_clean.groupby(by='designated').mean('n_count').reset_index()
n2=n1.rename(columns={'designated':'lineage'})

time_n=pd.merge(n2,earlytime,how='left',on=['lineage'])
time_n.to_csv('time_N.csv')
```

```python
import numpy as np
des_n['p_agree']=np.where(des_n['designated']==des_n['pango'],'yes','no')
des_n['u_agree']=np.where(des_n['designated']==des_n['usher'],'yes','no')

```

```python
exlist= relation['taxon'].loc[(relation['designated']!=relation['usher'])&
                            (relation['designated']!=relation['pango'])&
                            (relation['pango']==relation['usher'])].tolist()
des2=des_n[des_n.taxon.isin(exlist)==False].drop(columns=['Unnamed: 0'])
des2.loc[des2['u_agree'].str.contains('no')]

des2.to_csv('volion_test.csv')
```

```python
p_nolist=des2.loc[des2['p_agree']=='no','n_count'].tolist()
p_yeslist=des2.loc[des2['p_agree']=='yes','n_count'].tolist()

u_nolist=des2.loc[des2['u_agree']=='no','n_count'].tolist()
u_yeslist=des2.loc[des2['u_agree']=='yes','n_count'].tolist()
```

```python
from random import sample 
from scipy import stats
```

## PangoLEARN decision Tree N content T-test result!

```python
num=100
for _ in range(num):
    p_yessample=sample(p_yeslist,299)
    print(stats.ttest_ind(p_nolist, p_yessample, equal_var = False))
    
#all significant 
```

## UShER N content T-test result!

```python
num=100
for _ in range(num):
    u_yessample=sample(u_yeslist,10208)
    print(stats.ttest_ind(u_nolist, u_yessample, equal_var = False))
    
#Ttest_indResult(statistic=1.69567213294207, pvalue=0.08996327668914332)
```

# What happened after May?

```python
import pandas as pd
```

```python
dt_time=pd.read_csv('p_time_acc.csv',index_col=0)
rf_time=pd.read_csv('rf_time_acc.csv',index_col=0)
u_time=pd.read_csv('u_time_acc.csv',index_col=0)
```

```python
dt_may=dt_time.loc[dt_time['earliest_date']>='2021-05-01'].rename(columns={'accuracy':'dt_accuracy'})
rf_may=rf_time.loc[rf_time['earliest_date']>='2021-05-01'].rename(columns={'accuracy':'rf_accuracy'})
u_may=u_time.loc[u_time['earliest_date']>='2021-05-01'].rename(columns={'accuracy':'u_accuracy'})
u_may
```

```python
may1=pd.merge(dt_may,rf_may,on=['lineage','earliest_date'],how='outer')

may2=pd.merge(u_may,may1,on=['lineage','earliest_date'],how='outer')
may_sum=may2.drop(['lineage','earliest_date'], axis=1)
may_sum.to_csv('may_acc.csv')
```
